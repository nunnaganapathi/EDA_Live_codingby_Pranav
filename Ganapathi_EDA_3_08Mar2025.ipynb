```python
'''
Question 5:- Correlatin analysis

'''
```

```python
plt.figure(figsize=(7,3))
sns.heatmap(df.corr().round(3),annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
'''
Conclusions->
(1) "mileage" and "year" have a strong negative correlation
(2) "mileage" and "price" have a moderate negative correlation
(3) "mpg" and "tax" have a strong positive correlation
(4) "price" and "mpg" have a moderate negative correlation
(5) "price" and "engineSize" have a moderate positive correlation
(6) "Year" and "price" have a moderate positive correlation
(4) "mileage" and "tax" have a moderate negative correlation
'''
```

```python
'''
Question 7:- Plot a boxplot for "year" and "price"
'''
'''
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df['year'],y=df['price'],data=df,palette='Set1')
plt.title('Boxplot for Year and Price')
plt.show()
```

```python
'''
Conclusions:-

(1) Year 1970 seems to be the outlier
(2) From the year 1997 till 2010, the data seems to be consistent
(3)After 2010, there has been linear increase in the price till 2020
(4) The number of outliers from 2014 in the data has been significantly increased.
'''

```
