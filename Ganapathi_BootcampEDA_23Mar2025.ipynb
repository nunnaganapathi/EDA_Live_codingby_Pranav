```python
#Import important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Scatter plot
#Creating the data
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,100,86,103,87,94,78,77,85,86]
#Setting figure size
plt.figure(figsize=(5,3))

#Plot
plt.scatter(x,y,color='red')
plt.title('Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

```python
#Line Chart
#Creating the data
year=[1920,1930,1940,1950,1960,1970,1980,1990,2020,2010,2020]
unempoloyment_rate=[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3,10.2]

#Plot
plt.figure(figsize=(5,3))
plt.plot(year, unempoloyment_rate, color='green', marker='o')
plt.title('Line Chart')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
data={'java':20,
      'python':15,
      'sql':60}
#Setting figure size in Inches
plt.figure(figsize=(5,3))

#Plot
plt.bar(data.keys(),data.values(),color='blue', width=0.6)
plt.title('Popularity of Programming Langues')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
#Reading an inbuilt datase called 'tips'
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Count Plot
sns.countplot(x='day',data=tips)
#We can show Smokers and Tips
```

```python
#Count Plot
#sns.countplot(x='day',data=tips)
#We can show Smokers and Tips
sns.countplot(x='smoker',data=tips)
```

```python
#Pie Chart
#creating the data
labels=['Python','Java','C','C++']
sizes=[100,200,500,50]
plt.figure(figsize=(10,10))
plt.pie(sizes,labels=labels,autopct='%.2f%%')
plt.legend(title='Programming Languages')
plt.show()
```

```python
#Heat Map -> Colored Table
#Creating Data
data=np.random.randint(low =1,
                    high =100,
                    size=(10,10))
print('Raw Data :',data)
```

```python
#Heat Map
sns.heatmap(data,annot=True,cmap='Blues')
```
